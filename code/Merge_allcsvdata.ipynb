#合併txt

```python
txt_output_path = drive_path + 'txt_files/ICD9_1094.txt'
ype_path = drive_path + 'txt_files/icd9_txt/'

med_type = ["內科","家醫科","外科","洗腎科","皮膚科","耳鼻喉科","神經內科","神經外科","骨科","婦產科","眼科","精神科"] #泌尿科
txt_files = [type_path+t+'.txt' for t in med_type] 

cnt = 0
with open(txt_output_path, "w") as outfile:
    for f in txt_files:
        with open(f, "r") as infile:
            outfile.write(infile.read())
            cnt += 1
        outfile.write('\n')
outfile.close()
```

##確認txt順序 == csv順序

```python
TXT_icd9 = []
with open(txt_output_path, 'r') as R:
  lines = R.readlines()
  for icd9 in lines:
    TXT_icd9.append(icd9.strip())
R.close()

CSV_icd9 = []
with open(output_path, newline='') as csvfile:
  rows = csv.DictReader(csvfile)
  for row in rows:
    key_icd = row['ICD-9-CM 2001年版'].strip()
    CSV_icd9.append(key_icd)

same = 0
for i, j in zip(TXT_icd9, CSV_icd9) :
  if i != j : print(j)
  else : same+=1
same
```
