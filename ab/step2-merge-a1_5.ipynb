# Merge

## build a dic for current icd9 

```python
TYPE = "外科重跑"
TYPE_NUM = "48"
```

```python
# ================================ modify ================================

cur_icd9_txt_path = drive_path + 'txt_files/科別/' + TYPE + '_' + TYPE_NUM + '.txt'

# =========================================================================

# get the disease_id_list based on IM_icd9.txt
disease_id_list = []
check_cnt = 0
with open(cur_icd9_txt_path , "r") as file1:
    for line in file1.readlines():
        icd9_str = str(line)
        if icd9_str[0].isdigit():
            disease_str = icd9_str.strip()
            disease_id_list.append(disease_str)  
            check_cnt +=1
print(check_cnt)
```

# Merge & output

```python
di = 0
for disease in disease_id_list[di:check_cnt]:
  # Initialize 
  icd9_dic = {}
  # Specify the folderename
  output_folder = folder_path + '暫存/' + TYPE + '暫存/ab_merged/'
  # Specify the filename
  file_name = ""
  if len(disease)==3 : file_name = disease.strip()+".csv"
  else : file_name= disease[:disease.find('.')]+disease[disease.find('.')+1]+'.csv'

  # Read files{filname} from ab_1, ab_2, ... ab_5
  for i in range(1,6):
    input_folder = folder_path +  '暫存/' + TYPE + '暫存/ab_' + str(i) + '1/'
    with open(input_folder + file_name, newline='') as csvfile:
      rows = csv.DictReader(csvfile, delimiter=",")
      for row in rows:
        icd9_tmp = row['icd9'].strip()
        icd9_cnt = int(row['count'])
        if (icd9_dic.get(icd9_tmp) == None): icd9_dic[icd9_tmp] = icd9_cnt
        else : icd9_dic[icd9_tmp] += icd9_cnt
  
  #df = pd.DataFrame({'icd9': after_cnt[:,0], 'count':map(int,after_cnt[:,1])})
  df = pd.DataFrame(icd9_dic.items(), columns=['icd9', 'count'])
  df.count = pd.to_numeric(df.count, errors='coerce')
  df = df.sort_values(by=['count'],ascending=False)
  df.to_csv(output_folder + file_name,index=0)
  
  # check di
  print(di)
  di += 1
```

## check member in disease_id_list are all merged

```python
merged_file_cnt = 0
for disease in disease_id_list[0:check_cnt]:
  # Specify the filename
  file_name = ""
  if len(disease)==3 : file_name = disease.strip()+".csv"
  else : file_name= disease[:disease.find('.')]+disease[disease.find('.')+1]+'.csv'
  filepath = ("/content/drive/Shareddrives/專題/ab/暫存/" + TYPE + "暫存/ab_merged/" + file_name)
  
  if not os.path.isfile(filepath):
    print(file_name)
  else:
    merged_file_cnt+=1
if (merged_file_cnt == check_cnt) : print("Having same count as the check_cnt")
else : print("Inconsistent of merged_file_cnt & check_cnt")
```
